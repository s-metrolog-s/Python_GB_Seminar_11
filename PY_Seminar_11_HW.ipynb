# Дана функция:
### *f(x) = -12x^4*sin(cos(x))-18x^3+5x^2+10x-30**
1. Определить корни
2. Найти интервалы, на которых функция возрастает
3. Найти интервалы, на которых функция убывает
4. Построить график
5. Вычислить вершину
6. Определить промежутки, на котором f > 0
7. Определить промежутки, на котором f < 0

```python
from sympy import *
from sympy.abc import x
from sympy.plotting import plot

x = symbols('x')

# вывод корней через solveset
solveset(-12*x**4*sin(cos(x))-18*x**3+5*x**2+10*x-30, x, domain = Reals)
```

```python
# Построим график на отрезке [-100, 100]
import matplotlib.pyplot as plt

x = [(-12*x**4*sin(cos(x))-18*x**3+5*x**2+10*x-30) for x in range(-100,101)]
y = [y for y in range(-100,101)]
plt.plot(y, x)
plt.grid(axis = 'both')
plt.title('График функции')
plt.show()
```

```python
# Построим график на отрезке [-50, 50]
x = [(-12*x**4*sin(cos(x))-18*x**3+5*x**2+10*x-30) for x in range(-50,51)]
y = [y for y in range(-50,51)]
plt.plot(y, x, color = 'g')
plt.title('График функции')
plt.grid(axis = 'both')
plt.show()
```
